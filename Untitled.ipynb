```python
import os
import pandas as pd
import datetime
```

```python
# Replace this directory with your respective folder
home_dir = 'D:/IU/git_repo/COVID-19/csse_covid_19_data/csse_covid_19_daily_reports/'

files = os.listdir(home_dir)

files = [f for f in files if f not in {'.gitignore', 'README.md'}]
```

```python
df2 = []
for f in files:
    df1 = pd.read_csv(home_dir+f)
    df1['Date'] = f
    df2.append(df1)
```

```python
final = pd.concat(df2).reset_index(drop=True)

final.head(2)
```

```python
final['Date'] = final['Date'].apply(lambda x: x.split('.')[0])
```

```python
final['Date'].unique()
```

```python
final['Date'] = final['Date'].apply(lambda x: datetime.datetime.strptime(x, '%m-%d-%Y'))
```

```python
final['Province_State'] = final['Province_State'].fillna('')
```

```python
final2 = final[['Province_State', 'Country_Region', 'Confirmed', 'Date']]
```

```python
final2 = final2.groupby(['Date', 'Country_Region', 'Province_State'], as_index=False).sum()
```

```python
final2 = final2.sort_values(by=['Country_Region', 'Province_State', 'Date']).reset_index(drop=True)
```

```python
dates = final2.pop('Date')
```

```python
confirmed = final2.groupby(['Country_Region', 'Province_State']).diff()
```

```python
final2['Confirmed_daily'] = confirmed
```

```python
final2['Date'] = dates
```

```python
final2['Confirmed_daily'] = final2['Confirmed_daily'].fillna(0)
```

```python
final2 = final2.groupby(['Country_Region', 'Province_State'], as_index=False)['Confirmed'].diff()
```
